```python
# making candle sticks pattern 
from matplotlib.dates import DateFormatter, date2num, WeekdayLocator, DayLocator, MONDAY

infy_reset = infy.loc['2020-02':'2020-03'].reset_index()

infy_reset['date_ax'] = infy_reset['Date'].apply(lambda date: date2num(date))
infy_values = [tuple(vals) for vals in infy_reset[['date_ax','Open','High','Low','Close']].values]

mondays = WeekdayLocator(MONDAY)
alldays = DayLocator()
weekFormatter = DateFormatter('%b %d')
dayFormatter = DateFormatter('%d')

fig, ax = plt.subplots(figsize=(15,7))
candlestick_ohlc(ax, infy_values, width = 0.6, colorup ='green',colordown='red')
```

```python
# Daily Percentage Change
wipro['returns'] = (wipro['Close']/wipro['Close'].shift(1)) - 1
```

```python
wipro.head()
```

```python
techm['returns'] = (techm['Close']/techm['Close'].shift(1)) - 1
infy['returns'] = (infy['Close']/infy['Close'].shift(1)) - 1
```

```python
techm.head()
```

```python
infy.head()
```

```python
#Volatility
wipro['returns'].hist(bins = 50)
```

```python
techm['returns'].hist(bins=50)
```

```python
infy['returns'].hist(bins=50)
```

```python
wipro['returns'].hist(bins=100,label='Wipro',figsize=(15,7))
techm['returns'].hist(bins=100,label='Techm')
infy['returns'].hist(bins=100,label='Infy')
plt.legend()
```

```python
# kernal density estimation
wipro['returns'].plot(kind='kde',label='Wipro',figsize=(15,7))
techm['returns'].plot(kind='kde',label='Techm')
infy['returns'].plot(kind='kde',label='Infy')
plt.legend()
```

```python
# Making Box Plots
box_df = pd.concat([wipro['returns'],techm['returns'],infy['returns']],axis = 1)
box_df.columns = ['Wipro Returns','Techm Returns', 'Infy Returns']
box_df.plot(kind='box',figsize=(15,7))
```
